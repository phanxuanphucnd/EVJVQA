## Analysis

```python
with open('./data/private-test/evjvqa_private_test-desc-lang-qtype.json', 'r', encoding='utf-8') as f:
    test_data = json.load(f)
```

```python
annotations = test_data['annotations']

lang_dict = {}
qtype_dict = {
    'en': {},
    'vi': {},
    'ja': {}
}

for anno in annotations:
    if anno['language'] in lang_dict:
        lang_dict[anno['language']] += 1
    else:
        lang_dict[anno['language']] = 1
```

```python
lang_dict
```

```python
for anno in annotations:
    if anno['question_type'] in qtype_dict[anno['language']]:
        qtype_dict[anno['language']][anno['question_type']] += 1
    else:
        qtype_dict[anno['language']][anno['question_type']] = 1
```
